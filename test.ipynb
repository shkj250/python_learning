```python
import pandas as pd

df1 = pd.DataFrame({"a":[1,2,3], "b":[10,20,30]})
df2 = pd.DataFrame([
    {"a":1,"b":10},
    {"a":2,"b":20},
    {"a":3,"b":30},
])
print(df1.shape)
print(df1.columns)
print(df1.dtypes)
print(df1.head())
print(df1.describe())
print(df1.info())
print("-----")
print(df2)
```

```python
print(df1["a"] ) 
print("-----")               # 单列 Series
print(df1[["a","b"]])  
print("-----")               # 多列
print(df1.loc[0,"a"])    
print("-----")               # 行标签+列标签
print(df1.iloc[0,0])     
print("-----")               # 按整数位置
print(df1[df1["a"] > 1])      
print("-----")               # 布尔过滤
print(df1.query("a > 1 and b < 30"))
```
